```python
from typing import List
import math

# this number is optimized for one journal papaer page
lines = 46

content = [
    
    # 2004 #
    #"\\emph{HCONE$_I$}~(2004)~\\cite{hcone_2004}", # not in selection criteria
    "OLA (2004)~\\cite{ola_2004}",
    #"\\emph{S-Match}~(2004)~\\cite{s_match_2004}", # not in selection criteria
    "\\emph{ASCO}~(2004)~\\cite{DBLP:conf/iceis/LeDG04}", 
    "!bRiMOM~(2004)~\\cite{DBLP:conf/awcc/TangLLW04}",
    
    # 2005 #
    "\\emph{MoA$_I$}~(2005)~\\cite{moa_2005}",
    "oMap (2005)~\\cite{omap_2005}",
    "CROSI~(2005)~\\cite{crosi_2005}",
    "!b\\emph{Mongiello \\& Totaro}~(2005)~\\cite{DBLP:conf/ecweb/MongielloT05}",
    "!b\\emph{Aleksovski \\& Klein}~(2005)~\\cite{DBLP:conf/kcap/AleksovskiK05}",
    
    # 2006 #
    "OWL-Ctx~(2006)~\\cite{owl_ctx_2006}",
    #"RiMOM~(2006)~\\cite{rimom_2006}", # already included in 2005 (non-OAEI version)
    "AUTOMS~(2006)~\\cite{automs_2006}",
    "DSSim~(2006)~\\cite{dssim_2006}",
    "HMatch~(2006)~\\cite{hmatch_2006}",
    "\\emph{Aleksovski et al.}~(2006)~\\cite{DBLP:conf/ekaw/AleksovskiKKH06,DBLP:conf/semweb/AleksovskiKH06}",
    "\\emph{Park et al.}~(2006)~\\cite{DBLP:conf/iiwas/ParkKLB06,DBLP:journals/ijecommerce/ParkK07}",
    "\\emph{Alasoud et al.}~(2006)~\\cite{DBLP:conf/odbis/AlasoudHS06}",
    "\\emph{Sen et al.}~(2006)~\\cite{DBLP:conf/odbis/SenSS06}",
    "\\emph{Reynaud \\& Safar}~(2006)~\\cite{DBLP:conf/semweb/ReynaudS06}",
    "\\emph{Abolhassani et al.}~(2006)~\\cite{DBLP:journals/webology/AbolhassaniHH06}",
    "\\emph{Chen et al.}~(2006)~\\cite{DBLP:conf/wetice/ChenTL06}",
    "!b\\emph{iMapper}~(2006)~\\cite{DBLP:journals/dke/SuG06}",
    "!bontoDNA~(2006)~\\cite{DBLP:journals/ets/KiuL06}",
    "!b\\emph{Nagy et al.}~(2006)~\\cite{DBLP:conf/aaaifs/NagyVM06}",
    "!b\\emph{ACAOM}~(2006)~\\cite{DBLP:conf/bncod/WangGWZ06,DBLP:conf/webist/WangGWLL06}",
    "!b\\emph{Trojahn et al.}~(2006)~\\cite{DBLP:conf/eumas/SantosMQV06}",
    "!b\\emph{Wang et al.}~(2006)~\\cite{DBLP:conf/ksem/WangWZSD06}",
    "!b\\emph{Kim et al}~(2006)~\\cite{DBLP:conf/semweb/KimPBL06}",
    "!b\\emph{Wang et al.}~(2006)~\\cite{DBLP:conf/smc/WangWD06}",
    
    # 2007 #
    "ASMOV~(2007)~\\cite{asmov_2007}",
    #"\\emph{ILIADS$_I$}~(2007)~\\cite{ilidas_2007}", # not in selection criteria
    "SEMA~(2007)~\\cite{sema_2007}",
    "X-SOM~(2007)~\\cite{xsom_2007}",
    "\\emph{iG-Match}~(2007)~\\cite{DBLP:conf/ssd/HessIC07}",
    "\\emph{Tan \& Lambrix}~(2007)~\\cite{DBLP:conf/semweb/TanL07}",
    "!b\\emph{Trojahn et al.}~(2007)~\\cite{DBLP:conf/argmas/SantosQV07}",
    "!b\\emph{PROMPT-V}~(2007)~\\cite{DBLP:conf/iih-msp/JinFLLC07}",
    "!b\\emph{Jin et al.}~(2007)~\\cite{DBLP:conf/iita/JinFLLC07}",
    "!b\\emph{IAOM}~(2007)~\\cite{DBLP:conf/lsms/WuW07}",
    "!b\\emph{Sen et al.}~(2007)~\\cite{DBLP:conf/otm/SenFS07}",
    "!b\\emph{UFOme}~(2007)~\\cite{DBLP:conf/swap/PirroT07}",
    
    # 2008 #
    "MapPSO~(2008)~\\cite{mappso_2008}",
    "\\emph{Alasoud et al.}~(2008)~\\cite{DBLP:conf/ismis/AlasoudHS08}",
    "\\emph{Jeong-Woo et al.}~(2008)~\\cite{DBLP:conf/webi/SonPP08}",
    #"\\emph{MatchPLanner}~(2008)~\\cite{matchplanner_2008}", # not in selection criteria
    #"\\emph{SeMap}~(2008)~\\cite{semap_2008}", # not in selection criteria
    "\\emph{e-CMS}~(2008)~\\cite{DBLP:journals/jbcs/LachtimMC09}",
    "!b\\emph{Kaza \\& Chen}~(2008)~\\cite{DBLP:journals/dss/KazaC08}",
    "!b\\emph{Trojahn et al.}~(2008)~\\cite{DBLP:journals/ijmso/SantosQV08,DBLP:journals/jods/SantosMQV08,DBLP:conf/argmas/SantosQV08}",
    "!b\\emph{Ichise}~(2008)~\\cite{DBLP:conf/ACISicis/Ichise08}",
    "!b\\emph{Cardoso et al.}~(2008)~\\cite{DBLP:conf/atal/CardosoTO08}",
    "!b\\emph{Zhang et al.}~(2008)~\\cite{DBLP:conf/cscwd/ZhangDZ08}",
    "!b\\emph{OMIE}~(2008)~\\cite{DBLP:conf/dils/BouzeghoubET08}",
    "!b\\emph{Fatemi et al.}~(2008)~\\cite{DBLP:conf/fomi/FatemiSA08}",
    "!b\\emph{Wang et al.}~(2008)~\\cite{DBLP:conf/iri/WangLC08}",
    "!b\\emph{SECCO}~(2008)~\\cite{DBLP:conf/kes/PirroRT08}",
    "!b\\emph{Lera et al.}~(2008)~\\cite{DBLP:conf/swws/LeraJP08}",
    
    # 2009 #
    "Agreement Maker~(2009)~\\cite{agreementmaker_2009}",
    "\\emph{Eckert et al.}~(2009)~\\cite{DBLP:conf/esws/EckertMS09}",
    "\\emph{Zhong et al.}~(2009)~\\cite{DBLP:conf/sigmod/ZhongLLXTZP09}",
    "!b\\emph{Xia et al.}~(2009)~\\cite{DBLP:conf/fskd/XiaZHT09}",
    "!b\\emph{Fern{\\'{a}}ndez et al.}~(2009)~\\cite{DBLP:conf/prima/FernandezVL09}",
    
    
    # 2010 #
    "Eff2Match~(2010)~\\cite{eff2match_2010}",
    "\\emph{Mascardi et al.}~(2010)~\\cite{mascardi_2010}", # Automatic Ontology Matching via Upper Ontologies: A Systematic Evaluation
    "NBJLM~(2010)~\\cite{nbjlm_2010}",
    "\\emph{ontoMATCH}~(2010)~\\cite{DBLP:conf/interaction/Lu10}",
    "\\emph{IROM}~(2010)~\\cite{DBLP:conf/samt/SergiehU10}",
    # UFOme not added again (UFOme: An ontology mapping system with strategy prediction capabilities)
    "!b\\emph{Cheatham}~(2010)~\\cite{DBLP:conf/cts/Cheatham10}",
    "!b\\emph{Wang et al.}~(2010)~\\cite{DBLP:conf/ksem/WangLB10}",
    "!b\\emph{SOCOM}~(2010)~\\cite{DBLP:conf/www/FuBO10}",
    
    # 2011 #
    "CSA~(2011)~\\cite{CSA_2011}",
    "LogMap~(2011)~\\cite{logmap_2011}",
    "MaasMatch~(2011)~\\cite{maasmatch_2011}",
    #"\\emph{MOMIS}~(2011)~\\cite{momis_2011}", # not in selection criteria
    "OMReasoner~(2011)~\\cite{omreasoner_2011}",
    "Optima~(2011)~\\cite{optima_2011}",
    "YAM++~(2011)~\\cite{yam++_2011}",
    "\\emph{Lin \& Krizhanovsky}~(2011)~\\cite{DBLP:conf/rcdl/LinK11}",
    # XMap++ (conference paper)
    "\\emph{Sadaqat et al.}~(2011)~\\cite{DBLP:conf/fskd/JanLAA11}",
    "\\emph{Thayasivam \& Doshi}~(2011)~\\cite{DBLP:conf/semco/ThayasivamD11}",
    
    # 2012 #
    "\\emph{Vaccari et al.}~(2012)~\\cite{DBLP:journals/geoinformatica/VaccariSPBM12}", 
    "\\emph{Liu et al.}~(2012)~\\cite{DBLP:journals/ijautcomp/LiuYZWH12}",
    "\\emph{Acampora et al.}~(2012)~\\cite{DBLP:journals/ijis/AcamporaLSV12}",
    "\\emph{OARS}~(2012)~\\cite{DBLP:journals/tsmc/JanLAMQ12}",
    # MapPSO journal
    "\\emph{Fern{\\'{a}}ndez et al.}~(2012)~\\cite{DBLP:conf/ic3k/FernandezMV12}",
    "\\emph{FuzzyAlign}~(2012)~\\cite{DBLP:conf/ic3k/FernandezVML12}",
    "\\emph{OACLAI}~(2012)~\\cite{DBLP:conf/icws/LiaoLZXWS12}",
    "\\emph{Song et al.}~(2012)~\\cite{DBLP:conf/kes/SongZC12}",
    
    # 2013 #
    "\\emph{Gulic et al.}~(2013)~\\cite{DBLP:conf/icist/GulicMV13}",
    # sphere journal article
    "\\emph{MAPSOM}~(2013)~\\cite{DBLP:conf/jist/JirkovskyI13}",
    # xmap ++ paper "Ontology alignment using artificial neural network for large-scale ontologies"
    "\\emph{Acampora et al.}~(2013)~\\cite{DBLP:journals/isci/AcamporaLV13,DBLP:conf/fuzzIEEE/AcamporaKLV13}",
    "AML~(2013)~\\cite{aml_2013}",
    "XMap~(2013)~\\cite{xmap_2013}",
    "SPHeRe~(2013)~\\cite{sphere_2013}",
    "ServOMap~(2013)~\\cite{servomap_2013}",
    
    # 2014 #
    # servomap journal article
    "\\emph{UFOM}~(2014)~\\cite{DBLP:conf/iri/ZhangPP14}",
    "\\emph{Todorov et al.}~(2014)~\\cite{DBLP:conf/semweb/TodorovHG13}",
    "\\emph{Xue et al.}~(2014)~\\cite{DBLP:journals/soco/XueWH14,DBLP:journals/cai/XueWHH14,DBLP:journals/eswa/XueWR14}",
    "\\emph{Jaiboonlue et al.}~(2014)~\\cite{DBLP:conf/jist/JaiboonlueNCLRS14}",
    "AOT/AOTL~(2014)~\\cite{aotl_2014}",
    "InsMT/InsMTL~(2014)~\\cite{insmtl_2014}",
    
    # 2015 #
    "ServOMBI~(2015)~\\cite{servombi_2015}",
    "DKP-AOM~(2015)~\\cite{DBLP:conf/semweb/Fahad15}",
    "\\emph{Kiren \& Shoaib}~(2015)~\\cite{DBLP:journals/aslib/KirenS15}",
    "\\emph{Nguyen \& Conrad}~(2015)~\\cite{DBLP:conf/ic3k/Nguyen015}",
    "\\emph{Wang}~(2015)~\\cite{DBLP:conf/ksem/Wang15a}",
    "\\emph{Xue et al.}~(2015)~\\cite{DBLP:journals/ai/XueW15,DBLP:journals/iajit/XueWH15,DBLP:journals/jis/XueW15,DBLP:conf/cis/XueLTZR15,DBLP:conf/iih-msp/Xue15}",
    "\\emph{Benaissa et al.}~(2015)~\\cite{DBLP:conf/ciia/BenaissaK15}",
    
    # 2016 #
    "ALIN~(2016)~\\cite{alin_2016}",
    "CroLOM~(2016)~\\cite{crolom_2016}", 
    "CroMatcher~(2016)~\\cite{cromatcher_2016}", # journal article also available
    "\\emph{OMI-DL}~(2016)~\\cite{DBLP:journals/tsc/LiuCLBWW16}",
    "\\emph{Anam et al.}~(2016)~\\cite{DBLP:conf/acsc/AnamKKL16}",
    "\\emph{Xie et al.}~(2016)~\\cite{DBLP:conf/aina/XieCSC16}",
    "\\emph{Mountasser et al.}~(2016)~\\cite{DBLP:conf/iiwas/MountasserOF16}",
    "\\emph{Idoudi et al.}~(2016)~\\cite{DBLP:journals/ijcisys/IdoudiESH16}",
    "\\emph{Xue et al.}~(2016)~\\cite{DBLP:conf/icgec/XueTZ16}",
    "\\emph{ALINSyn}~(2017)~\\cite{DBLP:conf/icgec/XueTZ16}",
    
    # 2017 #
    "KEPLER~(2017)~\\cite{kepler_2017}",
    "ONTMAT~(2017)~\\cite{ontmat_2017}",
    "\\emph{Xue et al.}~(2017)~\\cite{DBLP:journals/jihmsp/Xue17,DBLP:journals/kais/XueP17,DBLP:journals/kbs/XueL17}",
    # ALINSyn is interactive
    "\\emph{He et al.}~(2017)~\\cite{DBLP:journals/jihmsp/HeXZ17}",
    
    # 2018 #
    "SANOM~(2018)~\\cite{sanom_2018}",
    "EVOCROS~(2018)~\\cite{evocros_2018}",
    "FCA-Map~(2018)~\\cite{fcamapx_2018}",
    # skipped (interactive) [paper, 2018] Interactive ontology matching based on partial reference alignment 
    "\\emph{Ochieng \& Kyanda}~(2018)~\\cite{DBLP:journals/dpd/OchiengK18}",
    "\\emph{Roussille et al.}~(2018)~\\cite{DBLP:conf/ekaw/RoussilleMTT18}",
    "\\emph{Vennesland et al.}~(2018)~\\cite{DBLP:conf/ercimdl/VenneslandA18,DBLP:conf/ic3k/VenneslandGWNS18}",
    "\\emph{Refoufi \& Benarab}~(2018)~\\cite{DBLP:conf/medprai/RefoufiB18}",
    "\\emph{Kolyvakis et al.}~(2018)~\\cite{DBLP:conf/naacl/KolyvakisKK18}",
    "\\emph{Bulygin et al.}~(2018)~\\cite{DBLP:conf/rcdl/Bulygin18}",
    "\\emph{Kachroudi \& Yahia}~(2018)~\\cite{DBLP:journals/jodsn/KachroudiY18}",

    # 2019 #
    "ONTMAT1~(2019)~\\cite{ontmat1_2019}",
    "Lily~(2020)~\\cite{lily_2020}",
    "\\emph{WeGO++}~(2019)~\\cite{DBLP:journals/soco/RajeswariKV19}",
    "\\emph{Bulygin \& Stupnikov}~(2019)~\\cite{DBLP:conf/rcdl/BulyginS19}",
    # sanom 2019 journal article
    "\\emph{Biniz \& Fakir}~(2019)~\\cite{DBLP:journals/iajit/BinizF19}",
    "\\emph{Xue \& Chen}~(2019)~\\cite{DBLP:journals/memetic/XueC19}",
    
    
    # 2020 #
    # sanom hobbit
    # alin journal article
    "\\emph{Ibrahim et al.}~(2020)~\\cite{DBLP:conf/webi/IbrahimF0J20}",
    "\\emph{Real et al.}~(2020)~\\cite{real_using_2020}",
    "\\emph{Xue \& Chen}~(2020)~\\cite{DBLP:journals/sensors/XueC20}",
    
    # 2021 #
    "\\emph{Lv et al.}~(2021)~\\cite{DBLP:journals/access/LvJL21}",
    "\\emph{Zhu et al.}~(2021)~\\cite{DBLP:journals/wicomm/ZhuXJR21}",
    "\\emph{Xue et al.}~(2021)~\\cite{DBLP:journals/complexity/XueYJTMZ21}",
    
]

def get_column_content(content: List[str], line_count: int) -> List[List[str]]:
    result = []
    position = 0
    current_list = []
    for entry in content:
        current_list.append(entry)
        position += 1
        if position == line_count:
            result.append(current_list)
            current_list = []
            position = 0
    
    if len(current_list) < line_count:
        result.append(current_list)
    return result


def generate_table(content: List[str], line_count: int, caption: str, label: str) -> str:
    column_content = get_column_content(content, line_count=line_count)
    num_content_columns = len(column_content)
    
    ls = ""
    for i in range(0, num_content_columns):
        ls += "l"
    
    header = """% This table is generated by notebook wordnet_table.ipynb
\\begin{table*}[]
\\centering
\\begin{tabular}{ll}
\\toprule
\\textbf{Knowledge Source} & \\textbf{Used by System} """
    
    
    header +=  "\\\\ \n\midrule\n"
    
    content = "\\multicolumn{1}{c}{WordNet} & \\begin{tabular}{" + ls + "}"
    

    for line in range(0, line_count):
        
        for column_nr in range(0, num_content_columns):
            entry = ""
            try:
                entry = column_content[column_nr][line]
            except IndexError:
                pass
            
            if entry.startswith("!b"):
                entry = entry[2:len(entry)]
                # \\begin{tabular}[c]{@{}l@{}}
                entry = "\\begin{color}{blue}" + entry + "\\end{color}" #\end{tabular}
            
            content += entry
            
            if column_nr != num_content_columns - 1:
                # not the last column
                content += " & "
            else:
                content += "\\\\\n"
    
    
    footer = "\\end{tabular}\n\\\\\\bottomrule\n\\end{tabular}\n\\caption{" + caption + "}\n\\label{" + label + "}\n\\end{table*}\n%End of generated table." 
    
    return header + content + footer

get_column_content(content=content, line_count=lines)
print(generate_table(content=content, line_count=lines, caption="Matching systems using WordNet. Referenced is the first documented usage by the matching system. Systems that did not participate in the OAEI \\begin{color}{blue}at some point in time\\end{color} are italicized. Ontology integration systems are indicated by a subscript $I$. \\begin{color}{blue}Named systems are referred to using their system name.\\end{color}", label="tab:wn_usage"))
```
