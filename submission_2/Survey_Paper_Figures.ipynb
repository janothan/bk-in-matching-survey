We now plot the same figure divided in two parts for better readability in a paper.

```python
def plot_systems_time_frame(from_incl: int, to_incl: int, width: int, height: int) -> None:
    fig_size = plt.rcParams["figure.figsize"]
    fig_size[0] = width
    fig_size[1] = height


    fig, ax = plt.subplots()
    ax.margins(y=0.0)

    bar_height = 0.25
    added_systems = 0
    system_labels = []
    
    for i,key in enumerate(data):
        first_participation = 9999
        for dates in data[key]:
            local_min = min(dates)
            if local_min < first_participation:
                first_participation = local_min
        
        # do not add the system if it first participated before from_incl
        if first_participation < from_incl:
            continue
        
        # do not add the system if it first participated after to_incl
        if first_participation > to_incl:
            continue
        
        added_systems += 1
        system_labels.append(key)
        
        for dates in data[key]:
            begin = min(dates) - 0.25
            end = max(dates) + 0.25
            #print(begin)
            #print(end)
            #print(i)
            ax.broken_barh([(begin, end-begin)], (added_systems - bar_height/2, bar_height), facecolors='black', label=key)

    ax.grid()
    ax.set_xlabel("Years")
    ax.set_ylabel("Matching Systems")
    ax.set_xticks(range(2004,2022))
    ax.set_yticks(range(added_systems + 1))
    ylabels = list(system_labels)
    ylabels.insert(0, "")
    ax.set_yticklabels(ylabels)

    plt.savefig(f"matching_systems_{from_incl}_to_{to_incl}.png", bbox_inches='tight')
    plt.show()


plot_systems_time_frame(from_incl=2012, to_incl=2021, width=15, height=25)
plot_systems_time_frame(from_incl=2000, to_incl=2011, width=15, height=17)
```

## Intermediate Functions
The cell below contains reusable functions to plot the same style of figure for different data points.
Go to the next cells for the actual generation of figures.

```python
from typing import Dict
from enum import Enum


class DomainType(Enum):
    """An enumeration for the BK type."""
    GENERAL = "g"
    DOMAIN_SPECIFIC = "d"

class Dataset():
    """A class representing a single external dataset."""
    def __init__(self, name, domain):
        self.name = name
        self.domain = domain


def render_timeline(data_dictionary: Dict[Dataset, Dict[int, int]], lower_bound: int, upper_bound: int, file_name: str, y_axis_text: str) -> None:
    """The figure is saved and shown.
        
        lower_bound : int
            The beginning year.
        upper_bound : int
            The ending year, including!
    
    """
    
    # setup the plot
    fig, ax = plt.subplots()
    fig_size = plt.rcParams["figure.figsize"]
    fig_size[0] = 20
    fig_size[1] = 8
    ax.set_xticks(range(lower_bound, upper_bound+1))

    # we need to avoid that matplotlib repeats colors
    #colors = plt.cm.Spectral(np.linspace(0,1,30))
    ax.set_prop_cycle('color', plt.get_cmap('tab20').colors)
    
    for bk, bk_data in data_dictionary.items():
        x_data = []
        y_data = []
        for year in range(lower_bound, upper_bound+1):
            x_data.append(year)
            if year in bk_data:
                y_data.append(bk_data[year])
            else:
                y_data.append(0)
        ax.plot(x_data, y_data, label=bk.name, alpha=1, linewidth = 3)
        #ax.fill_between(x_data, 0, y_data)

    ax.set_xlabel("Years")
    ax.set_ylabel(y_axis_text)
    ax.grid()
    ax.legend()
    fig.savefig(file_name, bbox_inches='tight')
    plt.show()

    
def render_cumulated_usage(data_dictionary: Dict[Dataset, Dict[int, int]], lower_bound: int, upper_bound: int, file_name: str) -> None:

    system_count = []

    for bk, bk_data in data_dictionary.items():
        count = 0
        for _, frequency in bk_data.items():
            count += frequency
        system_count.append((bk.name.replace(" ", "\n"), count))

    fig, ax = plt.subplots()
    fig_size = plt.rcParams["figure.figsize"]
    fig_size[0] = 20
    fig_size[1] = 8

    # sort descendingly
    system_count = sorted(system_count, key=lambda tup: tup[1], reverse=True)
    system_count

    x_data = [x[0] for x in system_count]
    y_data = [x[1] for x in system_count]

    ax.bar(x_data, y_data, color="black")
    ax.set_xlabel("Background Knowledge Sources") # Sorted Descendingly According to Cumultated Usage
    ax.set_ylabel("Cumulated Usage Count")
    ax.grid()
    fig.savefig(file_name, bbox_inches='tight')
    plt.show()

    
def render_cumulated_usage_per_year_with_domain_split(data_dictionary: Dict[Dataset, Dict[int, int]], lower_bound: int, upper_bound: int, file_name: str) -> None:
    fig, ax = plt.subplots()
    fig_size = plt.rcParams["figure.figsize"]
    fig_size[0] = 20
    fig_size[1] = 8
    width = 0.5

    x_data_domain = []
    y_data_domain = []
    x_data_general = []
    y_data_general = []

    domain_dict = {}
    general_dict = {}

    for year in range(lower_bound, upper_bound + 1):
        x_data.append(year)
        for bk, bk_data in data_dictionary.items():
            if year in bk_data:
                if bk.domain is None:
                    print("ERROR: Domain must not be none. BK Dataset: " + bk.name)
                    continue
                if bk.domain == DomainType.DOMAIN_SPECIFIC:
                    if year in domain_dict:
                        domain_dict[year] = domain_dict[year] + bk_data[year]
                    else:
                        domain_dict[year] = bk_data[year]
                elif bk.domain == DomainType.GENERAL:
                    if year in general_dict:
                        general_dict[year] = general_dict[year] + bk_data[year]
                    else:
                        general_dict[year] = bk_data[year]
                else:
                    print("ERROR: Unknown domain: " + str(bk.domain))

    for year in range(lower_bound, upper_bound + 1):
        x_data_domain.append(year)
        if year in domain_dict:
            y_data_domain.append(domain_dict[year])
        else:
            y_data_domain.append(0)

    for year in range(lower_bound, upper_bound + 1):
        x_data_general.append(year)
        if year in general_dict:
            y_data_general.append(general_dict[year])
        else:
            y_data_general.append(0)

    general_bar = ax.bar(x=x_data_general, height=y_data_general, width=width, color="black")
    general_bar.set_label("General Knowledge")
    domain_bar = ax.bar(x=x_data_domain, height=y_data_domain, width=width, bottom=y_data_general, color="grey")
    domain_bar.set_label("Domain Knowledge")

    ax.grid()
    ax.set_xlabel("Years")
    ax.set_xticks(range(lower_bound, upper_bound+1))
    
    # determine maximum upper bound
    maximum = 0
    for x in range(0, len(y_data_general)):
        usage_in_year = y_data_general[x] + y_data_domain[x]
        if usage_in_year > maximum:
            maximum = usage_in_year
    
    ax.set_yticks(range(0, maximum + maximum % 2, 2))
    ax.legend()
    fig.savefig(file_name, bbox_inches='tight')
    plt.show()
```
